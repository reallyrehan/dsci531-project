```python
from glob import glob
from tqdm import tqdm
from pathlib import Path
import os
import sys
import subprocess
```

```python
# Ensure the directories to store audio files exsit
audios_dir = 'IEMOCAP_audios'
if not os.path.exists(audios_dir):
    os.makedirs(audios_dir)
```

```python
audios = set()
for session_num in range(1, 6):
    path = f'IEMOCAP_full_release/Session{session_num}/sentences/wav'
    audios.update(glob(path+'/**/*.wav', recursive=True))
print(f'found {len(audios)} audios')
```

```python
# Downsample audios to 16k and save all on the same folder
for audio in tqdm(audios):
    filename = audio.split('/')[-1].split('.')[0]
    command = f"ffmpeg -i {audio} -ab 256k -ac 2 -ar 16000 -vn {audios_dir}/{filename}.wav"
    subprocess.call(command, shell=True)
```
