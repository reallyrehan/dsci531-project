```python
import pandas as pd
from glob import glob
```

```python
files = glob('performance/*.csv')
```

```python
df = pd.DataFrame()

for file in files:
    tmp = pd.read_csv(file)
    modality = file.split('/')[-1].split('.')[0]
    tmp.index = [modality]
    df = pd.concat([df, tmp])
```

```python
df.sort_values('f1_macro', ascending=False)
```

```python
df.sort_values('roc_auc', ascending=False)
```

```python
df.sort_values('accuracy', ascending=False)
```
